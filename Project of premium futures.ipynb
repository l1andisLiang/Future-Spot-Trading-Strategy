```python
# 定义计算均值、标准差及置信区间的函数
def calculate_bounds(df, confidence_interval):
    mean_premium = df['premium_rate'].mean()
    std_premium = df['premium_rate'].std()
    upper_bound = mean_premium + confidence_interval * std_premium
    lower_bound = mean_premium - confidence_interval * std_premium
    return mean_premium, std_premium, upper_bound, lower_bound

confidence_interval_90 = 1.645  # 对应于90%的置信区间
confidence_interval_99 = 2.576  # 对应于99%的置信区间

# 计算四个数据集的90%上下边界
mean_btc_usdtspot, std_btc_usdtspot, upper_bound_btc_usdtspot, lower_bound_btc_usdtspot = calculate_bounds(btc_usdtspot_premium_df, confidence_interval_90)
mean_eth_usdtspot, std_eth_usdtspot, upper_bound_eth_usdtspot, lower_bound_eth_usdtspot = calculate_bounds(eth_usdtspot_premium_df, confidence_interval_90)
mean_btc_usd_spot, std_btc_usd_spot, upper_bound_btc_usd_spot, lower_bound_btc_usd_spot = calculate_bounds(btc_usd_spot_premium_df, confidence_interval_90)
mean_eth_usd_spot, std_eth_usd_spot, upper_bound_eth_usd_spot, lower_bound_eth_usd_spot = calculate_bounds(eth_usd_spot_premium_df, confidence_interval_90)

# 计算四个数据集的99%上下边界
_, _, stoploss_upper_bound_btc_usdtspot, stoploss_lower_bound_btc_usdtspot = calculate_bounds(btc_usdtspot_premium_df, confidence_interval_99)
_, _, stoploss_upper_bound_eth_usdtspot, stoploss_lower_bound_eth_usdtspot = calculate_bounds(eth_usdtspot_premium_df, confidence_interval_99)
_, _, stoploss_upper_bound_btc_usd_spot, stoploss_lower_bound_btc_usd_spot = calculate_bounds(btc_usd_spot_premium_df, confidence_interval_99)
_, _, stoploss_upper_bound_eth_usd_spot, stoploss_lower_bound_eth_usd_spot = calculate_bounds(eth_usd_spot_premium_df, confidence_interval_99)

# 打印结果
print(f"BTCUSDT Spot Premium Rate: mean={mean_btc_usdtspot}, std={std_btc_usdtspot}, upper_bound={upper_bound_btc_usdtspot}, lower_bound={lower_bound_btc_usdtspot}, stoploss_upper_bound={stoploss_upper_bound_btc_usdtspot}, stoploss_lower_bound={stoploss_lower_bound_btc_usdtspot}")
print(f"ETHUSDT Spot Premium Rate: mean={mean_eth_usdtspot}, std={std_eth_usdtspot}, upper_bound={upper_bound_eth_usdtspot}, lower_bound={lower_bound_eth_usdtspot}, stoploss_upper_bound={stoploss_upper_bound_eth_usdtspot}, stoploss_lower_bound={stoploss_lower_bound_eth_usdtspot}")
print(f"BTCUSD Spot Premium Rate: mean={mean_btc_usd_spot}, std={std_btc_usd_spot}, upper_bound={upper_bound_btc_usd_spot}, lower_bound={lower_bound_btc_usd_spot}, stoploss_upper_bound={stoploss_upper_bound_btc_usd_spot}, stoploss_lower_bound={stoploss_lower_bound_btc_usd_spot}")
print(f"ETHUSD Spot Premium Rate: mean={mean_eth_usd_spot}, std={std_eth_usd_spot}, upper_bound={upper_bound_eth_usd_spot}, lower_bound={lower_bound_eth_usd_spot}, stoploss_upper_bound={stoploss_upper_bound_eth_usd_spot}, stoploss_lower_bound={stoploss_lower_bound_eth_usd_spot}")
```

# E1

针对于btc_usd_spot_premium_df的数据集，有三列分别是close_spot, close_future, premium_rate，针对于BTCUSD Spot Premium Rate 90%的上边界，每次大于等于BTCUSD Spot Premium Rate 90%的上边界，那么就用100000/当时的close_spot，得到一个值存储为Open_position, 然后记录下当时的close_future值，之后在有Open_position的情况下，大于等于BTCUSD Spot Premium Rate90%的上边界不触发，直到小于等于BTCUSD Spot Premium Rate的1.2*mean，进行以下计算： (之前的close_future值-现在close_future值)/现在close_future值*Open_position+Open_position,存储值为 Close_position, 然后用Close_position*新的close_spot算出最终值，值存储为revenue,再进行以下计算：(revenue-100000)/100000, 值存储为profit
加上止损的条件，也就是stoploss,在有Open_position的情况下，当大于等于BTCUSD Spot Premium Rate99%的confidence interval触发止损，计算当前的 (之前的close_future值-现在close_future值)/现在close_future值*Open_position+Open_position,存储值为 Loss_position, 然后Loss_position*新的close_spot算出值，存储为revenue，再进行以下计算：(revenue-100000)/100000, 值存储为profit
关闭现在的position，重新回到none值。直到小于等于BTCUSD Spot Premium Rate 90%的上边界，继续开仓新的Open_position，然后就是以上策略反复进行


```python
import pandas as pd
import numpy as np

# 计算90%和99%置信区间的上边界和均值
mean_premium = btc_usd_spot_premium_df['premium_rate'].mean()
std_premium = btc_usd_spot_premium_df['premium_rate'].std()
confidence_interval_90 = 1.645  # 对应于90%的置信区间
confidence_interval_99 = 2.576  # 对应于99%的置信区间
upper_bound_90 = mean_premium + confidence_interval_90 * std_premium
upper_bound_99 = mean_premium + confidence_interval_99 * std_premium
mean_threshold = 1.2 * mean_premium

# 添加新列用于存储计算结果
btc_usd_spot_premium_df['Open_position'] = np.nan
btc_usd_spot_premium_df['close_future_at_open'] = np.nan
btc_usd_spot_premium_df['Close_position'] = np.nan
btc_usd_spot_premium_df['Loss_position'] = np.nan
btc_usd_spot_premium_df['revenue'] = np.nan
btc_usd_spot_premium_df['profit'] = np.nan
btc_usd_spot_premium_df['loss'] = np.nan

open_position = None
close_future_at_open = None

for i, row in btc_usd_spot_premium_df.iterrows():
    if row['premium_rate'] >= upper_bound_90 and open_position is None:
        open_position = 100000 / row['close_spot']
        close_future_at_open = row['close_future']
        btc_usd_spot_premium_df.at[i, 'Open_position'] = open_position
        btc_usd_spot_premium_df.at[i, 'close_future_at_open'] = close_future_at_open
        print(f"Opened position at index {i}: Open_position={open_position}, close_future_at_open={close_future_at_open}")
    elif open_position is not None and row['premium_rate'] <= mean_threshold:
        close_position = (close_future_at_open - row['close_future']) / row['close_future'] * open_position + open_position
        revenue = close_position * row['close_spot']
        profit = (revenue - 100000) / 100000
        
        btc_usd_spot_premium_df.at[i, 'Close_position'] = close_position
        btc_usd_spot_premium_df.at[i, 'revenue'] = revenue
        btc_usd_spot_premium_df.at[i, 'profit'] = profit
        print(f"Closed position at index {i}: Close_position={close_position}, revenue={revenue}, profit={profit}")
        
        # Reset the positions
        open_position = None
        close_future_at_open = None
    elif open_position is not None and row['premium_rate'] >= upper_bound_99:
        loss_position = (close_future_at_open - row['close_future']) / row['close_future'] * open_position + open_position
        loss = loss_position * row['close_spot']
        profit = (loss - 100000) / 100000 
        
        btc_usd_spot_premium_df.at[i, 'Loss_position'] = loss_position
        btc_usd_spot_premium_df.at[i, 'revenue'] = loss
        btc_usd_spot_premium_df.at[i, 'profit'] = profit 
        print(f"Stoploss triggered at index {i}: Loss_position={loss_position}, revenue={loss}, profit={profit}")
        
        # Reset the positions
        open_position = None
        close_future_at_open = None

print(btc_usd_spot_premium_df[['close_spot', 'close_future', 'premium_rate', 'Open_position', 'Close_position', 'revenue', 'profit', 'Loss_position', 'loss']])

total_profit_btc_usd_spot_premium_df = (btc_usd_spot_premium_df['profit'].dropna() - 0.0006925).sum()/7
print(f"Total profit: {total_profit_btc_usd_spot_premium_df}")
```

加上止损的条件，也就是stoploss,在有Open_position的情况下，当小于等于BTCUSDT Spot Premium Rate99%的lower bound confidence interval，触发止损，计算方法:
(现在的close_future值-之前的close_future值)/之前的close_future值*100000+100000, 存储值为 loss_position, 然后用(loss_position/新的close_spot-Open_position)*新的close_spot，得到的值存储为revenue,再进行以下计算：(revenue-100000)/100000, 值存储为profit, 然后关闭现在的position，重新回到none值。直到大于等于BTCUSD Spot Premium Rate 90%的下边界，继续开仓新的Open_position，然后就是以上策略反复进行


```python
import pandas as pd
import numpy as np

# 计算90%和99%置信区间的上边界和均值
mean_premium = btc_usdtspot_premium_df['premium_rate'].mean()
std_premium = btc_usdtspot_premium_df['premium_rate'].std()
confidence_interval_90 = 1.645  # 对应于90%的置信区间
confidence_interval_99 = 2.576  # 对应于99%的置信区间
lower_bound_90 = mean_premium - confidence_interval_90 * std_premium
lower_bound_99 = mean_premium - confidence_interval_99 * std_premium
upper_bound_80_mean = 0.8 * mean_premium

# 添加新列用于存储计算结果
btc_usdtspot_premium_df['Open_position'] = np.nan
btc_usdtspot_premium_df['close_future_at_open'] = np.nan
btc_usdtspot_premium_df['Close_position'] = np.nan
btc_usdtspot_premium_df['Loss_position'] = np.nan
btc_usdtspot_premium_df['revenue'] = np.nan
btc_usdtspot_premium_df['profit'] = np.nan
btc_usdtspot_premium_df['loss'] = np.nan

open_position = None
close_future_at_open = None

for i, row in btc_usdtspot_premium_df.iterrows():
    if row['premium_rate'] <= lower_bound_90 and open_position is None:
        open_position = 100000 / row['close_spot']
        close_future_at_open = row['close_future']
        btc_usdtspot_premium_df.at[i, 'Open_position'] = open_position
        btc_usdtspot_premium_df.at[i, 'close_future_at_open'] = close_future_at_open
        print(f"Opened position at index {i}: Open_position={open_position}, close_future_at_open={close_future_at_open}")
    elif open_position is not None and row['premium_rate'] >= upper_bound_80_mean:
        close_position = (row['close_future'] - close_future_at_open) / close_future_at_open * 100000 + 100000
        revenue = (close_position / row['close_spot'] - open_position) * row['close_spot']
        profit = (revenue) / 100000
        
        btc_usdtspot_premium_df.at[i, 'Close_position'] = close_position
        btc_usdtspot_premium_df.at[i, 'revenue'] = revenue
        btc_usdtspot_premium_df.at[i, 'profit'] = profit
        print(f"Closed position at index {i}: Close_position={close_position}, revenue={revenue}, profit={profit}")
        
        # Reset the positions
        open_position = None
        close_future_at_open = None
    elif open_position is not None and row['premium_rate'] <= lower_bound_99:
        loss_position = (row['close_future'] - close_future_at_open) / close_future_at_open * 100000 + 100000
        revenue = (loss_position / row['close_spot'] - open_position) * row['close_spot']
        profit = (revenue) / 100000
        
        btc_usdtspot_premium_df.at[i, 'Loss_position'] = loss_position
        btc_usdtspot_premium_df.at[i, 'revenue'] = revenue
        btc_usdtspot_premium_df.at[i, 'profit'] = profit
        print(f"Stoploss triggered at index {i}: Loss_position={loss_position}, revenue={revenue}, profit={profit}")
        
        # Reset the positions
        open_position = None
        close_future_at_open = None


print(btc_usdtspot_premium_df[['close_spot', 'close_future', 'premium_rate', 'Open_position', 'Close_position', 'revenue', 'profit', 'Loss_position', 'loss']])

total_profit_btc_usdtspot_premium_df = (btc_usdtspot_premium_df['profit'].dropna() - 0.0006925).sum()/7
print(f"Total profit: {total_profit_btc_usdtspot_premium_df}")
```

## 开仓比例

```python
# 计算四个数据集的开仓比例

def calculate_kelly_ratio(df, dataset_name):
    profit_data = df['profit'].dropna()
    num_wins = (profit_data > 0).sum()
    num_losses = (profit_data < 0).sum()
    total_trades = num_wins + num_losses

    if total_trades == 0:
        return f"{dataset_name} 无交易数据"

    # 计算盈利和亏损概率
    P_win = num_wins / total_trades
    P_loss = num_losses / total_trades

    # 计算预期盈利和预期亏损
    E_win = profit_data[profit_data > 0].mean()
    E_loss = -profit_data[profit_data < 0].mean()  # 使用负值表示亏损

    # 计算开仓比例
    f = (P_win - (P_loss / (E_win / E_loss))) / 10

    return f"{dataset_name} 开仓比例: {f:.4f}"

kelly_ratios = []
datasets = [
    (btc_usdtspot_premium_df, "btc_usdtspot"),
    (btc_usd_spot_premium_df, "btc_usd_spot"),
    (eth_usd_spot_premium_df, "eth_usd_spot"),
    (eth_usdtspot_premium_df, "eth_usdtspot")
]

for df, name in datasets:
    kelly_ratios.append(calculate_kelly_ratio(df, name))

kelly_ratios
```
